```python
# Last_Purchase_Date    object chages the object into date:
#  First Changes the rename columns:
df.rename(columns={"Annual_Income_k$":"Annual_Income"},inplace=True)
df.head(5)
```

```python
#  Changes the datatypes:
df["Age_int"]=df["Age"].astype(int)
df
```

```python
df["new_Age"]=df["Age"].apply(lambda x:x*2)
df
```

```python
#  Data Agregating And grouping:
df.head()
Grouped_means=df.groupby("Gender")["Age"].mean()
print(Grouped_means)
```

```python
Groupd_sum=df.groupby(["Preferred_Category","Region"])["Age"].sum()
print(Groupd_sum)
```

```python
#  Aggregae multiple function:

Groupd_agg=df.groupby("Region")["Age"].agg(["mean","sum","count"])
print(Groupd_agg)
```

```python
#  Mergin and joining Dataframe:
df1=pd.DataFrame({"Key":['A','B','C'], "Value1":[1,2,3]})
df2=pd.DataFrame({"Key":["A","B","D"],"Value2":[4,5,6]})
```

```python
df1
```

```python
df2
```

```python
#  Merage the Dataframe df1 and df2  On the Key Columns:
merge_data=pd.merge(df1,df2,on="Key",how="outer")
merge_data
```

```python
merge_df=pd.DataFrame(merge_data)
merge_df["Value1"]=merge_df["Value1"].fillna(merge_df["Value1"].sum(),inplace=True)
merge_data
```
